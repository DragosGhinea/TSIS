```python
import numpy as np

def rle_to_mask(rle, height, width):
    # Step 1: build flat mask
    flat = []
    val = 0
    for run in rle:
        flat.extend([val] * run)
        val = 1 - val

    flat = np.array(flat, dtype=np.uint8)

    # Step 2: reshape to HxW (column-major)
    mask = flat.reshape((height, width), order="F")
    return mask
```
